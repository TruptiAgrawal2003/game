```python
users = {}
questions = [
    {
        "question": "What is the capital of France?",
        "choices": ["1. Paris", "2. London", "3. Rome", "4. Berlin"],
        "answer": "1"
    },
    {
        "question": "Which of the following is a programming language?",
        "choices": ["1. HTML", "2. CSS", "3. Python", "4. SQL"],
        "answer": "3"
    },
    {
        "question": "Who developed the theory of relativity?",
        "choices": ["1. Isaac Newton", "2. Albert Einstein", "3. Nikola Tesla", "4. Galileo Galilei"],
        "answer": "2"
    },
    {
        "question": "What is the capital of Japan?",
        "choices": ["1. Tokyo", "2. Seoul", "3. Beijing", "4. Bangkok"],
        "answer": "1"
    },
    {
        "question": "What gas do plants absorb from the atmosphere?",
        "choices": ["1. Oxygen", "2. Carbon Dioxide", "3. Nitrogen", "4. Hydrogen"],
        "answer": "2"
    }
]

# Functions
def user_exists(username):
    return username in users

def verify_user(username, password):
    return users.get(username) == password

def registration():
    username = input("Enter your name: ")
    if user_exists(username):
        print("Already registered.")
    else:
        password = input("Enter your password: ")
        users[username] = password
        print("Successfully Registered!")

def login():
    username = input("Enter your name: ")
    password = input("Enter your password: ")
    if verify_user(username, password):
        print("Successfully logged in!")
        return True
    else:
        print("Invalid username or password.")
        return False

def quiz_game():
    username = input("Enter your username: ").upper()
    print(f"Hi {username}")
    print("Do you want to play a short quiz game?")
    s = input("YES or NO: ").lower()
    while s == 'yes':
        score = 0
        count = 0
        for i, q in enumerate(questions):
            print(f"\nQuestion {i+1}: {q['question']}")
            print("Choices are:")
            for choice in q['choices']:
                print(choice)
            answer = input("Enter your choice (1-4): ")
            if answer in ["1", "2", "3", "4"]:
                if answer == q['answer']:
                    print("Correct Answer!")
                    score += 10
                    count += 1
                else:
                    print("Wrong Answer!")
            else:
                print("Invalid choice! Please choose a number between 1 and 4.")
        print(f"\nFinal Score: {score}")
        accuracy = (count / len(questions)) * 100
        print(f"Your Accuracy: {accuracy:.2f}%")
        if accuracy == 100:
            print("Excellent! You got all questions right!")
        elif accuracy >= 70:
            print("Great job! Keep practicing to improve!")
        elif accuracy >= 50:
            print("Good effort! You can do better with more practice.")
        else:
            print("Don't worry, keep practicing and you'll improve!")
        print("\nDo you want to play again?")
        s = input("YES or NO: ").lower()

def main():
    print("Choose an option:")
    print("1. Registration")
    print("2. Login")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        registration()
        print("Choose an option:")
        print("1. Login")
        choice = int(input("Enter your choice: "))
        if choice == 1:
            if login():
                print("Choose an option:")
                print("1. Quiz")
                print("2. Exit")
                choice = int(input("Enter your choice: "))
                if choice == 1:
                    quiz_game()
    elif choice == 2:
        if login():
            print("Choose an option:")
            print("1. Quiz")
            print("2. Exit")
            choice = int(input("Enter your choice: "))
            if choice == 1:
                quiz_game()
    else:
        print("Invalid choice. Exiting.")

main()

```

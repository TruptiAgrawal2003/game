```python
import os
QUESTIONS_FILE = 'questions.txt'
USERS_FILE = 'users.txt'
def load_questions():
    questions = []
    choices = []
    answers = []
    with open(QUESTIONS_FILE, 'r') as file:
        for line in file:
            data = line.strip().split('|')
            questions.append(data[0])
            choices.append(data[1:5])
            answers.append(data[5])
    return questions, choices, answers

def save_user(username, password):
    with open(USERS_FILE, 'a') as file:
        file.write(f"{username},{password}\n")


def user_exists(username):
    with open(USERS_FILE, 'r') as file:
        for line in file:
            stored_user, _ = line.strip().split(',')
            if stored_user == username:
                return True
    return False


def verify_user(username, password):
    with open(USERS_FILE, 'r') as file:
        for line in file:
            stored_user, stored_pass = line.strip().split(',')
            if stored_user == username and stored_pass == password:
                return True
    return False

def registration():
    username = input("Enter your name: ")
    if user_exists(username):
        print("Already registered.")
    else:
        password = input("Enter your password: ")
        save_user(username, password)
        print("Successfully Registered!")

def login():
    username = input("Enter your name: ")
    password = input("Enter your password: ")
    if verify_user(username, password):
        print("Successfully logged in!")
        return True
    else:
        print("Invalid username or password.")
        return False

def quiz_game():
    questions, choices, answers = load_questions()
    username = input("Enter your username: ").upper()
    print(f"Hi {username}")
    print("Do you want to play a short quiz game?")
    s = input("YES or NO: ").lower()
    while s == 'yes':
        score = 0
        count = 0
        for i in range(len(questions)):
            print(f"\nQuestion {i+1}: {questions[i]}")
            print("Choices are:")
            for option in choices[i]:
                print(option)
            answer = input("Enter your choice (1-4): ")
            if answer in ["1", "2", "3", "4"]:
                if answer == answers[i]:
                    print("Correct Answer!")
                    score += 10
                    count += 1
                else:
                    print("Wrong Answer!")
            else:
                print("Invalid choice! Please choose a number between 1 and 4.")
        print(f"\nFinal Score: {score}")
        accuracy = (count / len(questions)) * 100
        print(f"Your Accuracy: {accuracy:.2f}%")
        if accuracy == 100:
            print("Excellent! You got all questions right!")
        elif accuracy >= 70:
            print("Great job! Keep practicing to improve!")
        elif accuracy >= 50:
            print("Good effort! You can do better with more practice.")
        else:
            print("Don't worry, keep practicing and you'll improve!")
        print("\nDo you want to play again?")
        s = input("YES or NO: ").lower()

def main():
    if not os.path.exists(USERS_FILE):
        with open(USERS_FILE, 'w') as file:
            pass
    if not os.path.exists(QUESTIONS_FILE):
        with open(QUESTIONS_FILE, 'w') as file:
            
            file.write("What is the capital of France?|1. Paris|2. London|3. Rome|4. Berlin|1\n")
            file.write("Which of the following is a programming language?|1. HTML|2. CSS|3. Python|4. SQL|3\n")
            file.write("Who developed the theory of relativity?|1. Isaac Newton|2. Albert Einstein|3. Nikola Tesla|4. Galileo Galilei|2\n")
            file.write("What is the capital of Japan?|1. Tokyo|2. Seoul|3. Beijing|4. Bangkok|1\n")
            file.write("What gas do plants absorb from the atmosphere?|1. Oxygen|2. Carbon Dioxide|3. Nitrogen|4. Hydrogen|2\n")

    print("Choose an option:")
    print("1. Registration")
    print("2. Login")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        registration()
        print("Choose an option:")
        print("1. Login")
        choice = int(input("Enter your choice: "))
        if choice == 1:
            if login():
                print("Choose an option:")
                print("1. Quiz")
                print("2. Exit")
                choice = int(input("Enter your choice: "))
                if choice == 1:
                    quiz_game()
    elif choice == 2:
        if login():
            print("Choose an option:")
            print("1. Quiz")
            print("2. Exit")
            choice = int(input("Enter your choice: "))
            if choice == 1:
                quiz_game()
    else:
        print("Invalid choice. Exiting.")

main()
```
